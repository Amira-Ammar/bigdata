```python
nb_nan = data_20k['Unnamed: 10'].isna().sum()

print(nb_nan)
# la colonne unamed 10 contient des infos autre que des Nan, donc on supprime les Nan
# et on ne garde que les infos autre que Nan puis on fusionne cette colonne avec LIBELLE_EMPLOI_GRADE
```

```python
nb_nan = data_20k['Unnamed: 11'].isna().sum()

print(nb_nan)
# on a pas toutes les données en NA car le nombre diffère de shape data_20k
# donc on supprime les NA de la colonne puis on ajoute la valeur à LIBELLE_EMPLOI_GRADE
# pareil pour la colonne 'Unamed : 10'
```

```python
#data_20k = data_20k.dropna(subset=['Unnamed: 10'])
data_20k['Unnamed: 10'].fillna(value='', inplace=True)
#data_20k = data_20k.dropna(subset=['Unnamed: 11'])
data_20k['Unnamed: 11'].fillna(value='', inplace=True)
```

```python
data_20k['LIBELLE_EMPLOI_GRADE_'] = data_20k.apply(lambda row: str(row['LIBELLE_EMPLOI_GRADE']) + str(row['Unnamed: 11']) + str(row['Unnamed: 10']), axis=1)
data_20k = data_20k.drop(columns='LIBELLE_EMPLOI_GRADE')
data_20k = data_20k.drop(columns='Unnamed: 11')
data_20k = data_20k.drop(columns='Unnamed: 10')

data_20k.head(5)
```
